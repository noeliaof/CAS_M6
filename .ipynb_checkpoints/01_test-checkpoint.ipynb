```python
def plt_image_animation(frames, update_func):
    fig, ax = plt.subplots(figsize=(4,4))
    plt.axis('off')
    anim = animation.FuncAnimation(fig, update_func, frames=frames)
    video = anim.to_html5_video()
    html = display.HTML(video)
    display.display(html)
    plt.close()
```

```python
image_size = 128
```

```python
transformer = transforms.Compose([transforms.ToTensor(),
                                 torchvision.transforms.Resize(img_size),
                                  transforms.Normalize((0.5), (0.5)), #GrayScale
                                 ])
```

```python
class myDataset(Dataset):
    
    def __init__(self, X, transform):
        'Initialization'
        self.X = X
        self.transform=transform
        
    def __len__(self):
        'Denotes the total number of samples'
        return len(self.X)
    
 
    def __getitem__(self, index):
        'Generates one sample of data'
        # Select sample
        image = self.X[index]
        X = self.transform(image)
        
        return X
    
```

```python
train_dataset=myDataset(images,transformer)
```

```python
BATCH_SIZE = 64
train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE,shuffle=True)
```

```python
sample = next(iter(train_dataloader))
```

```python
def torch_tensor_to_plt(img):
    img = img.detach().numpy()[0]
    img = np.transpose(img, (1, 2, 0))
    return img 
```

```python
def imshow(inp, title=None):
    """imshow for Tensor."""
    inp = inp.numpy().transpose((1, 2, 0))
    inp = np.clip(inp, 0, 1)
    plt.imshow(inp)


# Get a batch of training data
sample = next(iter(train_dataloader))

# Make a grid from batch
output = torchvision.utils.make_grid(sample)
```
